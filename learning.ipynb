```python
import matplotlib.pyplot as plt

x= [0,1,2,3,4,5]
y= [0,1,4,6,8,16]

plt.title("Plain Graph")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[0,1,2,3,4,5,6]
y=[0,1,4,6,8,12,18]

fig, ax=plt.subplots()
ax.plot(x,y, marker='o', label='Data Points')
ax.set_title("Point Graph")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[0,1,4,6,8,10]
y=[0,1,2,3,4,5]

plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=['THURSDAY', 'FRIDAY', 'SATURDAY', 'SUNDAY']
y=[250,500,400,100]

plt.bar(x,y, fc='Green')
plt.title("BAR GRAPH")
plt.xlabel("DAYS")
plt.ylabel("PEOPLE VISITED")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [7, 8, 9, 10, 10, 12, 12, 12, 13, 14, 14, 15, 16, 16, 17, 18, 18, 19, 20, 20,
     21, 22, 23, 24, 25, 25, 26, 30, 32, 35, 36, 38, 40, 42, 44, 48, 50]

plt.hist(x, bins=10, color='#809de8')
plt.title('HISTROGRAM', color='#e0193a')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[20,30,40,50,60]
y=[200,300,400,500,600]
plt.scatter(x,y, fc='#ba1833')
plt.title("Scatter Graph", color='#90bdfc')
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars=["AUDI","LAMBO","BMW","TOYOTA"]
data=[10,25,32,30]

plt.pie(data,labels=cars,autopct='%1.1f%%',wedgeprops={'ec' : '#070e1a'}) ##ec=edgecolor
plt.title("PIE CHART", color='#2eb396')
plt.show()
```
