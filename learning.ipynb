```python
import matplotlib.pyplot as plt

x= [0,1,2,3,4,5]
y= [0,1,4,6,8,16]

plt.title("Plain Graph")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[0,1,2,3,4,5,6]
y=[0,1,4,6,8,12,18]

fig, ax=plt.subplots()
ax.plot(x,y, marker='o', label='Data Points')
ax.set_title("Point Graph")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
plt.show()
```
